## 3. Correlatie

We willen de samenhang van de verschillende variabelen bekijken. Als variabelen sterk gecorreleert met elkaar zijn betekend dit dat er een verband te stellen is tussen die variabelen en ze dus mogelijk afhankelijk van elkaar fungeren. 

```python
data.corr()
```

We tonen een klein overzicht van de variabelen die het sterkst gecorreleert zijn met elkaar. De variabelen die het sterkst met elkaar gecorreleert zijn, zoals bijvoorbeeld stm_progfh_gw_lwd_datum en stm_rapportage_jaar zeggen ons niet heel veel. Het is best logisch dat de prognose datum en het jaar waarin de rapportage gedaan is sterk met elkaar correleren. We zijn door deze correlaties dus niet perse veel wijzer geworden.

```python
data.corr().unstack().sort_values().drop_duplicates()
```
